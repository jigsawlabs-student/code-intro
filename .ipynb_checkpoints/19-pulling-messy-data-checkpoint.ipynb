```python
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)_per_capita'
tables = pd.read_html(url)

table = tables[1].iloc[1:, [0, 1, 4, 5]]
```

```python
table
```

```python
table.columns = ['country', 'region', 'gdp', 'year']
```

```python
table
```

```python
table['gdp'] = pd.to_numeric(table['gdp'], errors = 'coerce')
```

```python
table['year'] = pd.to_numeric(table['year'], errors = 'coerce')
```

```python
table['country'] = table['country'].str[:-1]
```

```python
country_names = list(table['country'])
```

```python
countries = [country_name.replace('\u202f', '') for country_name in country_names]
```

```python
table['country'] = countries
```

```python
table.to_csv('countries.csv', index = False)
```
